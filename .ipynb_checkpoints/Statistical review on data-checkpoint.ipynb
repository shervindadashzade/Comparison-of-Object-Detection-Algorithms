## Review our dataset 

#### Step #1 : Load our dataset

```python
import os
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
dataset_root = os.path.join('Dataset')
dataset_images_path = os.path.join(dataset_root,'JPGImages')
dataset_annotations = os.path.join(dataset_root,'FL_Keys_Coral-export.csv')
dataset_images = os.listdir(dataset_images_path)
```

```python
annotation_df = pd.read_csv(dataset_annotations)
annotation_df
```

```python
annotation_counted = annotation_df.groupby('image').count()
annotation_counted
```

```python
print(f'Images length {len(dataset_images)} and we have {len(annotation_counted)} images recorded on our dataset annotations')
```

```python
not_recorded = []
for x in dataset_images:
    if x not in annotation_counted.index:
        not_recorded.append(x)
        dataset_images.remove(x)
not_recorded
```

### Review the classes and thier counts

```python
classes_df = annotation_df.groupby('label').agg({'image':'count'}).rename(columns = {'image':'count'}).sort_values(by='count',ascending=False)
classes_df
```

```python
print(f'We have {len(classes_df)} classes which includes : {classes_df.index.values}')
```

```python
label_per_class_df = annotation_df.groupby(['image','label']).agg({'xmin':'count'}).rename(columns={'xmin':'count'})
label_per_class_df
```

### Visualizing our data

```python
labels = annotation_counted.index.values
classes = classes_df.index.values
c_count_per_name_dict = {}
for c in classes:
    c_count_per_name_dict[c] = []

for l in labels:
    for c in classes:
        if c in label_per_class_df.loc[l].index:
                c_count_per_name_dict[c].append(label_per_class_df.loc[(l,c)].values[0])
        else:
            c_count_per_name_dict[c].append(0)

c_count_per_name_dict
```

```python
df = pd.DataFrame()
```

```python
c_count_per_name_dict['image'] = labels
```

```python
df = pd.DataFrame(c_count_per_name_dict)
```

```python
df
```

```python
df.plot(x='image',kind='bar',stacked=False,title='Classes per image')
```
